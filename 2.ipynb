```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("AcademicPerformance.csv")
```

```python
print(df)
```

```python
df.isnull().sum()
```

```python
df['math score'] = df['math score'].ffill()
```

```python
# Visualization
import seaborn as sns
sns.boxplot(df['math score'])

# Outlier removal using IQR
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['math score'] < (Q1 - 1.5 * IQR)) | (df['math score'] > (Q3 + 1.5 * IQR)))]
```

```python
#not in practical
import seaborn as sns
import matplotlib.pyplot as plt

# Boxplot for math score
sns.boxplot(x=df['math score'])
plt.title("Boxplot for Math Score")
plt.show()

# Outlier removal using IQR for math score
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1

# Filter out outliers
df = df[~((df['math score'] < (Q1 - 1.5 * IQR)) | (df['math score'] > (Q3 + 1.5 * IQR)))]
```

```python
import numpy as np
df['Log_Math_Score'] = np.log(df['math score'] + 1)
```
